```python
# Imports
import pandas as pd
import numpy as np
import re
```

```python
def trace_filter_abf(array):
    pattern = r"\d{4}_\d{2}_\d{2}_\d{4}\.abf"
    filtered = [item for item in array if re.match(pattern+'.abf', str(item))]
    return np.array(filtered)
```
