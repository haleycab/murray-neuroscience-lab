```python
# Imports
import pandas as pd
import numpy as np
import re
```

```python
def trace_filter_abf(array):
    pattern = r"\d{4}_\d{2}_\d{2}_\d{4}\.abf"
    filtered = [item for item in array if re.match(pattern+'.abf', str(item))]
    return np.array(filtered)
```

### Names of each annotation sheet

```python
sheet_names_df = pd.read_csv('/Users/haleyoro/Desktop/murray-neuroscience-lab/Processed CSVs/sheet_names.csv', header=None)
sheet_names = sheet_names_df.iloc[:,0].values
sheet_names # names of each annotation sheet
```

### All abf files available in shared folder

```python
abfs = pd.read_csv('/Users/haleyoro/Desktop/murray-neuroscience-lab/Excel processor/all_abf_files.csv',header=None)
abfs = abfs[0].to_numpy()
abfs 
```

Check to see duplicates

```python
pattern = r"\d{4}_\d{2}_\d{2}_\d{4}\.abf"

abfs = [f for f in abfs if re.fullmatch(pattern, f)]

print(len(abfs)) 
print(len(set(abfs)))
```

All traces in each sheet are unique

```python
all_trace_names = []
for sheet in sheet_names:
    file_path = "/Users/haleyoro/Desktop/murray-neuroscience-lab/Processed CSVs/"+ sheet + '.csv'
    df_filtered = pd.read_csv(file_path)
    trace_names = df_filtered["Trace name"].unique().tolist()
    all_trace_names.extend(trace_names)

print(len(all_trace_names)==len(set(all_trace_names)))
print(len(all_trace_names))
```

Note: only 158 out of 794 abf files have annotations

```python
pattern = r"\d{4}_\d{2}_\d{2}_\d{4}\.abf"

missing_abfs = [f for f in missing_abfs if re.fullmatch(pattern, f)]

print(len(missing_abfs)==len(set(missing_abfs)))
print(len(missing_abfs))
```

Find which traces are missing in the downloaded folder:

```python
missing_abfs_base = [name.replace(".abf", "") for name in missing_abfs]

matches_missing = [name for name in all_trace_names if name in missing_abfs_base]

print(f"{len(matches_missing)} matches found:")
print(matches_missing)
```
